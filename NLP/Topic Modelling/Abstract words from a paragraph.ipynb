### Task: find top words appearing in the paragraph

#### Step
- Download the data 
- Transform words using count vectorizer
- Latent DirichAllocation 
- Get prob of each word LDA.components_
- Sort into probabilites
- Find the top words


```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AvinashTiwari/Artifical-intelligence/master/7_nlp_python/19_Latent_Dir/npr.csv')
```

```python
df.head(3)
```

```python
len(df)
```

```python
cv = CountVectorizer(max_df=0.9,min_df=2,stop_words='english')
```

```python
dtm = cv.fit_transform(df['Article'])
```

```python
LDA = LatentDirichletAllocation(n_components=7,random_state=42)
```

```python
LDA.fit(dtm)
```

```python
# grap highest prob of words per topic

single_topics = LDA.components_[0]
single_topics = single_topics.argsort()
print(cv.get_feature_names_out()[single_topics[-15:]])
```

```python
df['Article'][0]
```

```python
for index,topic in enumerate(LDA.components_):
    print(index,[cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print("\n\n")
```
